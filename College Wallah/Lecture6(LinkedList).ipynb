# Linked List


## Creating a linked list.

```python
# Class Node to create Node.

class Node:
    def __init__(self,data):
        self.data = data 
        self.next = None

a = Node(2)
b = Node(8)
c = Node(17)
d = Node(1)

a.next=b
b.next=c
c.next=d

head = a
```

## Implementing soem operations in linked list

```python
# Function to Traverse & print the linked list

def traverse_print(head):
    curr = head

    while curr!=None:
        print(curr.data,end=" -> ")
        curr = curr.next
    print("None")

traverse_print(head)
```

```python
# Inserting a new node in the Beginning

newNode = Node(0)
newNode.next =  head
head =  newNode

traverse_print(head)

```

```python
# Inserting a new node in the end

newNode = Node(2.8)

curr =  head

while curr.next != None:
    curr=curr.next

curr.next = newNode

traverse_print(head)
```

```python
# Inserting a new node at Kth index.
k = 2
newNode = Node(17.2)

curr = head

for _ in range(k-1):
    curr = curr.next

newNode.next =  curr.next
curr.next = newNode

traverse_print(head)

```

```python
# Delete the head node.

head = head.next

traverse_print(head)
```

```python
# Delete the last node.

curr = head

while curr.next.next!=None:
    curr=curr.next

curr.next=None

traverse_print(head)
```

```python
# Delete the node at Kth index.
k = 1

curr = head

for _ in range(k-1):

    curr=curr.next

curr.next = curr.next.next

traverse_print(head)

```

```python
# Search for a value in the linked list and returning the index and node.

def search(value):
    idx = 0
    curr = head

    while curr != None:
        if curr.data == value:
            return idx,curr
        
        curr = curr.next
        idx+=1
    return -1,None

value = 17  

idx , foundNode = search(value)

print(f"The value {value} is at index {idx}")
```

```python
# Concept of slow & fast pointer.
newNode = Node(17.2)

newNode.next = head
head = newNode

slow = head
fast = head

while fast.next!=None:
    slow=slow.next
    fast=fast.next.next

traverse_print(head)
print("Middle value: ",slow.data)

```
